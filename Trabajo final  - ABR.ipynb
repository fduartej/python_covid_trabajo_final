```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
departamentos = gpd.read_file('/content/per_admbnda_adm2_ign_20200714.shp')
departamentos['ADM1_ES_MAYUS']= departamentos['ADM1_ES'].str.upper()
departamentos.head()
```

```python
centros_vacunacion=pd.read_csv('/content/TB_CENTRO_VACUNACION.csv')
centros_vacunacion.head()
```

```python
ubigeos=pd.read_csv('/content/TB_UBIGEOS.csv')
ubigeos.head()
```

```python
lista_columunas=['ubigeo_reniec','ubigeo_inei','departamento_inei','provincia_inei','region','macroregion_inei','macroregion_minsa','iso_3166_2','fips','superficie','altitud','latitud','longitud']
ubigeos01=ubigeos.drop(lista_columunas,axis=1)
ubigeos01.head()
```

```python
centros_vacunacion01=centros_vacunacion.merge(ubigeos01,on="id_ubigeo",how="left")
centros_vacunacion01.head()
```

```python
#@title Eleccón de Departamento Preliminar { run: "auto" }
indice = 22 #@param {type:"slider", min:0, max:24, step:1}
seleccion_ubigeos = centros_vacunacion01[centros_vacunacion01['departamento']==lista_departamentos[indice]]
seleccion_geometria = gpd.GeoDataFrame(seleccion_ubigeos, geometry=gpd.points_from_xy(seleccion_ubigeos.longitud, seleccion_ubigeos.latitud))
seleccion_geometria01=seleccion_geometria[(seleccion_geometria['latitud']!=0)|(seleccion_geometria['longitud']!=0)]
eleccion=departamentos[departamentos['ADM1_ES_MAYUS']==lista_departamentos[indice]]
fig, ax = plt.subplots(figsize = (8,10))
eleccion.plot(ax = ax, color = 'White', edgecolor = 'grey')
seleccion_geometria01.plot(ax = ax, color = 'Red')
```
